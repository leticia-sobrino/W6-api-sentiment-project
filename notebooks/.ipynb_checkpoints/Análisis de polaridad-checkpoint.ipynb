### Polaridad por año
- Conseguir una funcion que meta el año que quiera y que le de la media de la polaridad de esos tweets

```python
year = list(collection.find({'year':{"$in": ["2019"]}}, {"content": 1, "_id" : 0}))
year
```

```python
def sentimentAnalysis(tweet):
    sia = SentimentIntensityAnalyzer()
    polarity = sia.polarity_scores(tweet)
    pol = polarity['pos']
    return pol
```

```python
pos = []
for i in year:
    pos.append(sentimentAnalysis(i['content']))
print(pos)    
```

```python
mean = sum(pos)/len(pos)
mean
```
